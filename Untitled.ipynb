```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import lstsq
from sklearn.neighbors import KNeighborsRegressor
```

```python
data = pd.read_csv('./data/wdbc.csv')
data.head()
```

```python
x = np.array(data['mean_radius'])
y = np.array(data['mean_concave_points'])
M = x[:, np.newaxis]**[0,1]
p, res, rnk, s = lstsq(M, y)
```

```python
plt.plot(x, y, 'o', label='data')
xx = np.linspace(0, 40, 101)
yy = p[0] + p[1]*xx
plt.plot(xx, yy, label='least squares fit, $y = a + bx$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(framealpha=1, shadow=True)
plt.grid(alpha=0.25)
plt.show()
```

```python
# Build our model.
knn = KNeighborsRegressor(n_neighbors=25)
X = pd.DataFrame(data['mean_radius'])
y = data.mean_concave_points
knn.fit(X, y)

# Set up our prediction line.
T = np.arange(5, 30, 0.1)[:, np.newaxis]

# Trailing underscores are a common convention for a prediction.
y_ = knn.predict(T)

plt.scatter(X, y, c='k', label='data')
plt.plot(T, y_, c='g', label='prediction')
plt.legend()
plt.title('K=10, Unweighted')
plt.show()
```
